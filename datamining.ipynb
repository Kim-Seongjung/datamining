```python
def findEvidence(list_key , name , source_path  ,show = False):
    wb = open_workbook(source_path)
    #get sheet
    all_sheet=wb.sheets()
    finding_sheet=all_sheet[0] #
    nrow=finding_sheet.nrows
    print '# row  : ', nrow
    
    returned_list_CTFinding=[]
    returned_list_CTResult=[]
    dict_ele={}
    #split row 
    for key in list_key:
        for patient_code in range(nrow):
            split_CTFinding=finding_sheet.row_values(patient_code)[3].split('\n') # CT Finding
            split_CTResult=finding_sheet.row_values(patient_code)[4].split('\n') # CT Finding

            for j in range(len(split_CTFinding)):
                if key in split_CTFinding[j]:
                    evid=split_CTFinding[j]
                    ele=[patient_code , key, name, evid ,'CTFinding'] #tuple
                    returned_list_CTFinding.append(ele)
                    
            for i in range(len(split_CTResult)):
                if key in split_CTResult[i]:
                    evid=split_CTResult[i]
                    ele = [patient_code ,key , name ,evid , 'CTResult']
                    returned_list_CTResult.append(ele)
                
                    
    if show == True:
        print 'CTFinding'
        n_ele = len(returned_list_CTFinding)
        print '############################# valid evidence :' , n_ele,'#############################' 
        for i in range(n_ele):
            print returned_list_CTFinding[i][1]+' : '+returned_list_CTFinding[i][3]
            #split_CTResut=finding_sheet.row_values(i)[4].split('\n') # CT Result
        print 'CTResult'
        n_ele = len(returned_list_CTResult)
        print '############################valid evidence :' , n_ele,'#############################' 
        for i in range(n_ele):
            print returned_list_CTResult[i][1]+' : '+returned_list_CTResult[i][3]

        return returned_list_CTFinding,returned_list_CTResult
    
```

```python
def checkPercent(path , start_row ):
    print "para 2 : real row number in excel Don't count zero"
    
    wb= open_workbook(path)
    all_sheet = wb.sheets()
    sheet= all_sheet[0]
    n_row=sheet.nrows
    n_col = sheet.ncols
    print n_row , n_col 
    #allocate dynamic list 
    count=[]
    for i in sheet.row_values(0):
        count.append(0)
        
    for i in range(start_row , n_row):
        for j in range(len(count)):
            if sheet.row_values(i)[j] != u'':
                count[j]+=1
                
    j=0
    for j in range(len(count)):
        
        print "col "+str(j)+' percentage : '+str(count[j])+'/'+str(n_row) , (count[j]*100 / n_row) ,'%'
    
```

```python
def copy_list(list_):
    returned_list=[]
    for ele in list_:
        returned_list.append(ele)
    return returned_list

def findPattern(p1 ,p2,p3  , list_source):
# p1 have highest priority
# list_source form = code , key , name , evid , CTFinding

    n_row = len(list_source)
    list_correct = []
    print '# of row ' , n_row 
    #print list_source[0]
    list_num_equivocal=[]
    list_equivocal=[]
    for i in range(n_row): 
        evid = list_source[i][3]
        #print evid
        if p1 in evid  and p2 =='': 
            list_correct.append(copy_list(list_source[i]))
            list_source[i]=''
        elif p1 in evid and p2 !='':
            if p2 in evid and p3 =='':
                list_correct.append(copy_list(list_source[i]))
                list_source[i]=''
            elif p2 in evid and p3 !='':
                if p3 in evid:
                    list_correct.append(copy_list(list_source[i]))
                    list_source[i]=''
    
    for ele in list_source:
        if ele!= '':
            list_equivocal.append(ele)
    return list_correct  , list_equivocal
```

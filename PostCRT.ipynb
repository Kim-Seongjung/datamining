```python
from  __future__ import unicode_literals
def excelDate2readDate(list_date , wb ): #wb is object 

    import xlrd
    import re
    p1=re.compile('\d+/+\d+/+\d{1,2}')
    p2=re.compile('\d+-+\d+-+\d{1,2}')
      
    for i,date in enumerate(list_date):
        try:
            m1=p1.match(str(date)) #type 이 datetime이면 검색이 안된다. 
            m2=p2.match(str(date))
            if m1:
                list_date[i]=m1.group()
                print i,list_date[i]
            elif m2:
                list_date[i]=m2.group()
                print i,list_date[i]
            elif date!='':
                raw_date=xlrd.xldate.xldate_as_datetime(date, wb.datemode)
                date=str(raw_date)

                g1=p1.match(date)
                g2=p2.match(date)
                if g1:
                    list_date[i]=g1.group()
                    print i,list_date[i]
                elif g2:
                    list_date[i]=g2.group()
                    print i,list_date[i]
                else:
                    print raw_date
        except UnicodeEncodeError as uee:
            list_date[i]=''
            continue
        except ValueError as ve:
            list_date[i]=''
            continue
    return list_date
```

```python
def getYYMMDDfromDate(dateList): #element has to be YY-MM-DD
    list_tuple_date=[]
    for i,ele in enumerate(dateList):
        if ele != '':
            list_tuple_date.append(ele.split('-'))
            
        else:
            list_tuple_date.append('')
            continue
    return list_tuple_date
```

```python
def extractTN(post_CRTs , list_disease ):
    import re
    list_T=[]
    list_N=[]
    for i,ele in enumerate(post_CRTs):
        try:
            disease=list_disease[int(ele[0])]
            patternT=re.compile('T+\d{1}')
            patternN=re.compile('N+\d{1}')
            mT=patternT.match(disease)
            
            if mT:
                str_=patternT.sub(mT.group()+' ' ,disease)
                T= str_.split()[0]
                equi_N=str_.split()[1]
                list_T.append((ele[1],T))
                mN=patternN.match(equi_N)
                if mN:
                    N=mN.group()
                    list_N.append((ele[1],N))
        except TypeError as te:
            continue

    return list_T ,list_N
            
        
        
```
